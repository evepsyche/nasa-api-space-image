```python
#escolhendo exibir so imagens
if(apod["media_type"] == "image"):


    #escolhendo exibir apenas imagens HD
    if("hdurl" in apod.keys()):


        #salvando nome da imagem
        title = d + "_" + apod["title"].replace(" ","_").replace(":","_") + ".jpg"


        #caminho do diretorio
        image_dir = "./imagens_espaco"




        #Pegando a imagem usando a função urllib
        urllib.request.urlretrieve(url = apod["hdurl"] , filename = os.path.join(image_dir,title))


        #informações da imagem


        if("title" in apod.keys()):
            print("Título: ",apod["title"])
            print("\n")
        if("explanation" in apod.keys()):
            print("Descrição: ",apod["explanation"])
            print("\n")
        if("hdurl" in apod.keys()):
            print("URL: ",apod["hdurl"])
            print("\n")

        #exibindo imagem
        display(Image(os.path.join(image_dir,title)))
```
